# LangChain: Evaluation

```python
from L4 import *
```

For creating Q&A data structure

```python
file = 'OutdoorClothingCatalog_1000.csv'
data = load_csv(file)
index = new_vector_db(file)
qa = new_qa(index, sep="<<<<>>>>>", verbose=True)
```

For creating new data samples

```python
examples = [
    {
        "query": "Do the Cozy Comfort Pullover Set\
        have side pockets?",
        "answer": "Yes"
    },
    {
        "query": "What collection is the Ultra-Lofty \
        850 Stretch Down Hooded Jacket from?",
        "answer": "The DownTek collection"
    }
]
examples += gen_examples(llm, data[:5])

with DebugContextManager(langchain):
    for example in examples[:3]:
        qa.run(example["query"])
```

For evaluating the predictions of the Q&A

```python
predictions = qa.batch(examples)
score = evaluate(examples, predictions)
score
```
